```python
#Round the parameters to an appropriate number of decimal places
def roundparams(param_array):
    p_array = []
    p_array.append(int(np.round(param_array[0],decimals=0)))
    p_array.append(np.round(param_array[1],decimals=2))
    p_array.append(np.round(param_array[2],decimals=2))
    p_array.append(np.round(param_array[3],decimals=2))
    p_array.append(np.round(param_array[4],decimals=2))
    p_array.append(np.round(param_array[5],decimals=2))
    p_array.append(np.round(param_array[6],decimals=2))
    p_array.append(np.round(param_array[7],decimals=2))
    p_array.append(param_array[8])
    p_array.append(np.round(param_array[9],decimals=1))
    return p_array


new_star1p = roundparams(star1param)
new_star2p = roundparams(star2param)
new_star3p = roundparams(star3param)
new_star4p = roundparams(star4param)
new_star5p = roundparams(star5param)
new_star6p = roundparams(star6param)
new_star7p = roundparams(star7param)
new_star8p = roundparams(star8param)
new_star9p = roundparams(star9param)
new_star10p = roundparams(star10param)
new_star11p = roundparams(star11param)

#rounded parameter table
ptable = Table(rows = [new_star1p,new_star2p,new_star3p,new_star4p,new_star5p,new_star6p,new_star7p,new_star8p,new_star9p,new_star10p,new_star11p])
ptable['col10'] = noms
ptable['col0'].name = 'TEFF'
ptable['col1'].name = 'LOGG'
ptable['col2'].name = 'LOGVMICRO'
ptable['col3'].name = 'M_H'
ptable['col4'].name = 'C_M'
ptable['col5'].name = 'N_M'
ptable['col6'].name = 'ALPHA_M'
ptable['col7'].name = 'col7'
ptable['col8'].name = 'col8'
ptable['col9'].name = 'SNR'
ptable['col10'].name = 'NAME'

ptable = ptable['NAME','TEFF','LOGG','LOGVMICRO','M_H','C_M','N_M','ALPHA_M','col7','col8','SNR']

ptable.show_in_browser(jsviewer=True)
```

```python
#Get rid of the bad pixels values

def badpixel(spec_array):
    for i in range(len(spec_array)):
        if spec_array[i] <= 0.75:
            spec_array[i] = 1.0
        if spec_array[i] >= 1.075:
            spec_array[i] = 1.0
    return spec_array

newstar1tab = badpixel(np.copy(star1tab['SPEC']))
newstar2tab = badpixel(np.copy(star2tab['SPEC']))
newstar3tab = badpixel(np.copy(star3tab['SPEC']))
newstar4tab = badpixel(np.copy(star4tab['SPEC']))
newstar5tab = badpixel(np.copy(star5tab['SPEC']))
newstar6tab = badpixel(np.copy(star6tab['SPEC']))
newstar7tab = badpixel(np.copy(star7tab['SPEC']))
newstar8tab = badpixel(np.copy(star8tab['SPEC']))
newstar9tab = badpixel(np.copy(star9tab['SPEC']))
newstar10tab = badpixel(np.copy(star10tab['SPEC']))
newstar11tab = badpixel(np.copy(star11tab['SPEC']))
```
